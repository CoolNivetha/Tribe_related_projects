```python
#now we import our dataset

event_data = pd.read_excel('Spring_Event_Feedback_Survey.xlsx')
event_data.head()
```

We commence our data cleaning process

```python
#below are some of the identified issues with our data
"""
Column error
removing rows with nan
expanding date column
expanding columns with multiple responses
dropping unwanted columns
"""
```

```python
#dropping unwanted columns
event_data.drop(columns= ['Event Feedback Survey','Unnamed: 9' ], inplace= True)
```

```python
event_data
```

```python
#renaming columns
extracted_row = list(event_data.loc[1])
extracted_row

extracted_row[0] = 'Gender'
#extracted_row

event_data.columns = extracted_row
```

```python
#dropping unwanted rows
event_data.drop(index=[0, 1, 17], inplace= True )
```

```python
#event_data.reset_index()
event_data.reset_index()
```

```python
event_data['Event Date'] = pd.to_datetime(event_data['Event Date'])
event_data.head()
```

```python
#expanding date columns to extract month and day

event_data['event_day'] = ''

event_data['event_day'] = event_data['Event Date'].dt.day
```

```python
#extracting event month

event_data['event_month'] = ''

event_data['event_month'] = event_data['Event Date'].dt.month_name()
event_data
```

Exploratory Data analytics

```python
#total counts of responses

data_desc = event_data.describe()
data_desc
```

```python
#gender distribution of respondents

gender_dist = event_data['Gender'].value_counts().reset_index(name= 'res_count')
gender_dist
```

```python
#no of respondents who have attended previous events
prev_event_att = event_data['Have you previously attended any events organized by Tribe?'].value_counts().reset_index(name= 'counts')
prev_event_att
```

```python
#preferred event day

day_pref = event_data['What day of the week are most convenient for you to attend events?'].str.split(',').explode().str.strip().value_counts().reset_index(name='counts')
day_pref
```

```python
#most preferred time of event

time_pref = event_data['During which time(s) of the day are you most likely to attend events?'].str.split(',').explode().str.strip().value_counts().reset_index(name= 'counts')
time_pref
```

```python
#gender specific events

event_pref = event_data['Would you like to see more gender focused events (e.g. female entrepreneurship, including discussions with female founders and women in business)'].value_counts().reset_index(name='counts')
event_pref
```

```python
event_ads = event_data['Do you feel that event advertisements are relevant to your interests and needs as a member of your gender group'].value_counts(). reset_index(name = 'counts')
event_ads
```

```python
#return 

prev_event_att_by_gender= event_data.groupby('Gender')['Have you previously attended any events organized by Tribe?'].value_counts().reset_index(name= 'counts')
prev_event_att_by_gender
```

```python
#preferred day by gender
#event_data.assign(
    #pref_days = event_data['What day of the week are most convenient for you to attend events?']
    #.str.split(',').explode('pref_days').str.strip())
```

```python
#preferred event day by gender
event_data_exploded = event_data.assign(
    pref_days=event_data['What day of the week are most convenient for you to attend events?']
    .str.split(',')
).explode('pref_days')

# here we remove any leading or trailing whitespace from the new column
event_data_exploded['pref_days'] = event_data_exploded['pref_days'].str.strip()
event_data_exploded

#to get the preferred event days of the week grouped by gender
pref_event_day_by_gender = event_data_exploded.groupby('Gender')['pref_days'].value_counts().reset_index(name= 'counts')
pref_event_day_by_gender 
```

```python
#most preferred event time by gender

pref_time = event_data.assign(
    pref_times = event_data['During which time(s) of the day are you most likely to attend events?'].str.split(',').explode('pref_times')
)
pref_time['pref_times'] = pref_time['pref_times'].str.strip()
pref_time

#now we count preferred tims for gender

pref_time_by_gender = pref_time.groupby('Gender')['pref_times'].value_counts().reset_index(name= 'counts')
pref_time_by_gender
```

```python
#event type preferrence by gender
event_type_pref_by_gender = event_data.groupby('Gender')[
    'Would you like to see more gender focused events (e.g. female entrepreneurship, including discussions with female founders and women in business)'].value_counts().reset_index(name = 'counts').sort_values(['Gender', 'counts'], ascending= [True, False])
event_type_pref_by_gender
```

```python
#relevance of event ads to gender
rel_ad_gen = event_data.groupby('Gender')[
    'Do you feel that event advertisements are relevant to your interests and needs as a member of your gender group'].value_counts().reset_index(name = 'counts')
rel_ad_gen
```

```python
day_pref.rename(columns={'What day of the week are most convenient for you to attend events?': 'week_day'}, inplace= True)
```

```python
def replace_dayname(val):
    if val == 'Sundays':
        return 'Sun'
    if val == 'Mondays':
        return 'Mon'
    if val == 'Tuesdays':
        return 'Tue'
    if val == 'Wednesdays':
        return 'Wed'
    if val == 'Thursdays':
        return 'Thu'
    if val == 'Fridays':
        return 'Fri'
    if val == 'Saturdays':
        return 'Sat'
    else:
        return val

day_pref['week_day'] = day_pref['week_day'].apply(replace_dayname)
day_pref
```

```python
plt.bar(day_pref['week_day'], day_pref['counts'])
plt.title('Most preferred day of event')
plt.xlabel('weekday')
plt.show()
```

```python
#most preferred event day
plt.plot(day_pref['week_day'], day_pref['counts'])
plt.title('Most preferred day of event')
plt.xlabel('weekday')
plt.show()
```

```python
test_df = pref_event_day_by_gender.pivot(index= 'pref_days', columns='Gender', values= 'counts')
test_df

x = np.arange(len(test_df.index))
#now we set the width of our bars
bar_width = 0.35

plt.bar(x -bar_width/2, test_df['Male'], width= bar_width, label = 'Male')
plt.bar(x + bar_width/2, test_df['Female'], width = bar_width, label = 'Female')
plt.title("Preferred event day by Gender")
plt.xlabel('Weekday')
plt.legend()
plt.show()
```
