```python
#now we import our dataset

event_data = pd.read_excel('Spring_Event_Feedback_Survey.xlsx')
event_data.head()
```

We commence our data cleaning process

```python
#below are some of the identified issues with our data
"""
Column error
removing rows with nan
expanding date column
expanding columns with multiple responses
dropping unwanted columns
"""
```

```python
#dropping unwanted columns
event_data.drop(columns= ['Event Feedback Survey','Unnamed: 9' ], inplace= True)
```

```python
event_data
```

```python
#renaming columns
extracted_row = list(event_data.loc[1])
extracted_row

extracted_row[0] = 'Gender'
#extracted_row

event_data.columns = extracted_row
```

```python
#dropping unwanted rows
event_data.drop(index=[0, 1, 17], inplace= True )
```

```python
#event_data.reset_index()
event_data.reset_index()
```

```python
event_data['Event Date'] = pd.to_datetime(event_data['Event Date'])
event_data.head()
```

```python
#expanding date columns to extract month and day

event_data['event_day'] = ''

event_data['event_day'] = event_data['Event Date'].dt.day
```

```python
#extracting event month

event_data['event_month'] = ''

event_data['event_month'] = event_data['Event Date'].dt.month_name()
event_data
```

Exploratory Data analytics

```python
#total counts of responses

data_desc = event_data.describe()
data_desc
```

```python
#gender distribution of respondents

gender_dist = event_data['Gender'].value_counts().reset_index(name= 'res_count')
gender_dist
```

```python
#no of respondents who have attended previous events
prev_event_att = event_data['Have you previously attended any events organized by Tribe?'].value_counts().reset_index(name= 'counts')
prev_event_att
```

```python
event_data.head()
```

```python
#preferred event day

day_pref = event_data['What day of the week are most convenient for you to attend events?'].str.split(',').explode().str.strip().value_counts().reset_index(name='counts')
day_pref
```

```python
#most preferred time of event

time_pref = event_data['During which time(s) of the day are you most likely to attend events?'].str.split(',').explode().str.strip().value_counts().reset_index(name= 'counts')
time_pref
```

```python
#gender specific events

event_pref = event_data['Would you like to see more gender focused events (e.g. female entrepreneurship, including discussions with female founders and women in business)'].value_counts().reset_index(name='counts')
event_pref
```

```python
event_ads = event_data['Do you feel that event advertisements are relevant to your interests and needs as a member of your gender group'].value_counts(). reset_index(name = 'counts')
event_ads
```

```python
#return 

gen_return = event_data.groupby('Gender')['Have you previously attended any events organized by Tribe?'].value_counts().reset_index(name= 'counts')
gen_return
```

```python
#preferred day by gender
gen_pref_day = event_data.pivot_table(columns='Gender')['What day of the week are most convenient for you to attend events?'].str.split(',').explode().str.strip().value_counts().reset_index(name='counts')
gen_pref_day
```
